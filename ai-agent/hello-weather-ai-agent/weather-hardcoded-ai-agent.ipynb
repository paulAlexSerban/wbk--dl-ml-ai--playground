```python

from openai import OpenAI
import os
from dotenv import load_dotenv

load_dotenv()

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

openai = OpenAI(api_key=OPENAI_API_KEY)
```

```python
def get_weather(city: str) -> str:
    if city == "California":
        return "Sunny"
    if city == "New York":
        return "Rainy"
    if city == "Texas":
        return "Hot"
    if city == "Florida":
        return "Humid"
```

```python
def generate_text(prompt:str, system_prompt: str, model:str="gpt-4.1", max_tokens:int=100) -> str:
    """
    Generate text using OpenAI's API.

    Args:
        prompt (str): The input prompt for text generation.
        system_prompt (str): The system prompt to guide the generation.
        model (str): The model to use for generation.
        max_tokens (int): The maximum number of tokens to generate.

    Returns:
        str: The generated text.
    """
    response = openai.chat.completions.create(
        model=model,
        messages=[
            {"role": "user", "content": prompt},
            {"role": "system", "content": system_prompt}
        ],
        max_tokens=max_tokens
    )
    return response.choices[0].message.content
```

```python
current_weather = get_weather("California")
```

```python
prompt = f"""
Should I take an umbrella today?
Current weather in California: {current_weather}
"""

system_prompt = "You are a helpful assistant that provides weather advice."

response = generate_text(prompt, system_prompt)
print(response)  # Output the response from the AI agent
```
