```python
import torch
from transformers import pipeline
```

```python
sentiment_classifer = pipeline("sentiment-analysis")
```

```python
sentiment_classifer("I love using transformers library!")
```

```python
# named entity recognition
ner_pipeline = pipeline("ner", model="dslim/bert-base-NER")
ner_pipeline("Hugging Face is a technology company based in New York and Paris")
```

```python
zeroshot_classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
sequence_to_classify = "One day I will see the world"
candidate_labels = ["travel", "cooking", "dancing"]
```

```python
zeroshot_classifier(sequence_to_classify, candidate_labels)
```
